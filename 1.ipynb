#          Seam carving algoritam


Na slici vidimo razliku između algoritma skaliranja i seam carving algoritma.
Slika je preuzeta od Andrew Palczewski sa sajta https://www.apharmony.com/software-sagacity/2014/07/seam-carving-the-smart-image-resizing-algorithm/.

Važnost piksela određuje se nekom funkcijom energije E. U zavisnosti od funkcije E, različiti delovi slike se mogu smatrati više ili manje važnim. Funkcija energije E koja se najčešće primenjuje je $$  \left \lvert \frac{\partial f}{\partial x} \right \rvert + \left \lvert \frac{\partial f}{\partial y} \right \rvert $$.
Ova funkcija daje meru promene na slici u oba koordinatna pravca. Ideja je da veća energija ukazuje na veću promenu, a manja na neke delove koji su u okolini veoma slični. Uklanjanjem piksela manje energije, manje se gubi detalja sa slike i slika izgleda prirodnije.
